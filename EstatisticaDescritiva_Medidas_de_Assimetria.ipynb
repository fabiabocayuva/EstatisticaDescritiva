As **medidas de assimetria** (como desvio padrão, variância, coeficiente de variação) nos permitem analisar o o **grau de dispersão** dos dados em torno de uma medida de posição (moda, mediana e média). 

Neste código se encontra um estudo inicial para entender como as medidas de assimetria podem ser utilizadas como ferramentas de análise e, em seguida, um exemplo simples aplicado a um dataframe.

Antes de começar, vamos carregar as bibliotecas necessárias e criar uma função que calcule automaticamente todas as estatíticas das bases de dados:

```python
# Importando bibliotecas
import matplotlib.pyplot as plt
import math 
import numpy as np
import seaborn as sns
import pandas as pd 
from scipy import stats
from scipy.stats import kurtosis, skew
%matplotlib inline

# Criando função para calcular estatísticas 
def calcula_estatisticas(data):
  mean = round(np.mean(data),2)
  mode = round(stats.mode(data,)[0][0],2)
  median = round(np.median(data),2)
  var = round(np.var(data),2)
  std = round(np.std(data),2)
  coef_pearson = round(stats.skew(data),2)
  coef_kurtosis = round(kurtosis(data),2)
  msg = print("Media: {}\nModa: {}\nMediana: {}\nVariância: {}\nDesvio Padrão: {}\nCoeficiente de Pearson: {}\nCoeficiente de Kurtosis: {}".format(mean, mode, median, var, std, coef_pearson, coef_kurtosis))
  return msg
```

# **Medidas de Assimetria como Ferramentas de Análise em Estatística**

## **Considere três amostras de dados simples s0, s1 e s2**

```python
s0 = [4,5,6,5,5,6,6,5,4,5,5,5]
s1 = [12,8,4,3,2,1,1]
s2 = [-3,-1,4,2.5,4,3]
```

### **Analisando a primeira amostra de dados s0**

```python
calcula_estatisticas(s0)
```

```python
fig = plt.figure(figsize=(15,3))
ax = sns.distplot(s0,rug=True, kde=True, hist=False)

ax.annotate('Mode', xy=(5,0.4), xytext=(4.7,0.7), arrowprops={'arrowstyle':'-'})
ax.annotate('Median', xy=(5,0.4), xytext=(5.3,0.7), arrowprops={'arrowstyle':'-'})
ax.annotate('Media', xy=(5,0.4), xytext=(5.0,0.9), arrowprops={'arrowstyle':'-'})
plt.show()
```

**Distribuição Simétrica**

Nesse conjunto de dados observamos uma distribuição bem próxima de simétrica, onde as medidas de tendência central (média, moda e mediana) têm valores *iguais* ou muito próximos. 

Além disso, tem-se o coeficiente de Pearson *próximo de zero*, sendo mais um indicativo de uma distribuição simétrica. O coeficiente de Kurtosis com valor *menor que 3* indica que o grau de achatamento da distribuição normal é *platicúrtica* (mais achatada que o platô).

### **Analisando a segunda amostra de dados s1**

```python
calcula_estatisticas(s1)
```

```python
fig = plt.figure(figsize=(15,3))
ax = sns.distplot(s1,rug=True, kde=True, hist=False)

ax.annotate('Mode', xy=(1,0.01), xytext=(1,0.03), arrowprops={'arrowstyle':'-'})
ax.annotate('Median', xy=(3,0.01), xytext=(2.5,0.04), arrowprops={'arrowstyle':'-'})
ax.annotate('Media', xy=(4.4,0.01), xytext=(4.5,0.04), arrowprops={'arrowstyle':'-'})
plt.show()
```

**Distribuição Assimétrica Positiva**

Aqui observamos uma concentração de dados maior para a esquerda, com o coeficiente de Pearson próximo de 1, indicando uma distribuição assimétrica positiva.

### **Analisando a terceira amostra de dados s2**

```python
calcula_estatisticas(s2)
```

```python
fig = plt.figure(figsize=(15,3))
ax = sns.distplot(s2,rug=True, kde=True, hist=False)

ax.annotate('Mode', xy=(1,0.01), xytext=(1,0.07), arrowprops={'arrowstyle':'-'})
ax.annotate('Median', xy=(3,0.01), xytext=(3,0.09), arrowprops={'arrowstyle':'-'})
ax.annotate('Media', xy=(4.4,0.01), xytext=(4.5,0.07), arrowprops={'arrowstyle':'-'})
plt.show()
```

**Distribuição Assimétrica Negativa**

Essa análise é bem parecida com a anterior, porém com um desvio de concentração de dados para a direita do gráfico.

Aqui observamos uma concentração de dados maior para a direita, com o coeficiente de Pearson próximo de -1, indicando uma distribuição assimétrica negativa.

## **Considere amostras criadas aleatoriamente**

### **Amostras seguindo uma distribuição normal**

```python
sample_normal = np.random.normal(1,0.5,500)
sample_normal[:10]
```

```python
calcula_estatisticas(sample_normal)
```

É obtida uma distribuição simétrica:

```python
fig = plt.figure(figsize=(12,4))
sns.distplot(sample_normal)
```

### **Amostra com relacionamento exponencial entre os dados**

**Gerando uma distribuição assimétrica positiva**

```python
exp = np.arange(2,len(sample_normal),2) # começando em 2, do mesmo tamanho da amostra de sample_normal, incremento 2 em 2
exp[:10]
```

```python
sample_assimetrica_positiva = [] 
for s,e in zip(sample_normal,exp):
  sample_assimetrica_positiva.append(s*e) 

sample_assimetrica_positiva[:10]
```

```python
calcula_estatisticas(sample_assimetrica_positiva)
```

```python
fig = plt.figure(figsize=(12,4))
sns.distplot(sample_assimetrica_positiva)
```

**Gerando uma distribuição assimétrica negativa**

```python
exp = (np.arange(2,len(sample_normal),2))*-1 # começando em 2, do mesmo tamanho da amostra de sample_normal, incremento 2 em 2
exp[:10]
```

```python
sample_assimetrica_negativa = [] 
for s,e in zip(sample_normal,exp):
  sample_assimetrica_negativa.append(s*e) 

sample_assimetrica_positiva[:10]
```

```python
fig = plt.figure(figsize=(12,4))
sns.distplot(sample_assimetrica_negativa)
```

# **Exemplo prático e simples com um Dataframe**

```python
# Carregando a base de dados
df = pd.read_csv("/content/drive/MyDrive/DataSciencedoZero/Colab Notebooks/cars.csv")

# Limpando os dados do atributo MSRP
df['MSRP'] = [x.replace('$', '') for x in df['MSRP']]
df['MSRP'] = [x.replace(',', '') for x in df['MSRP']]
# Transformando a coluna do tipo object para tipo numerico para poder aplicar as funções matemáticas
df['MSRP'] = pd.to_numeric(df['MSRP'], errors='coerce')
```

Analisando graficamente as distribuições dos dados de cada atributo do Dataframe

```python
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(16,12))

for i, columns in enumerate(df.select_dtypes(include=np.number).columns.tolist()):
  sns.distplot(df[columns],ax=axes[i//3,i%3])
```

Verificando a distribuição de probabilidade e a diferença entre a média e a mediana:

- Coeficiente de variação abaixo de 25% é um bom indicativo de que temos uma boa homegeneidade dos dados, enquanto que acima de 25% indica uma dispersão maior entre os dados.

```python
for i, column in enumerate(df.select_dtypes(include=np.number).columns.tolist()):
  media = round(df[column].mean(),2)
  std = df[column].std()
  cvar = round(std/media*100,2)
  median = df[column].median()
  print("Atributo: {}, Coeficiente de Variação: {}%, Média: {}, Mediana: {}".format(column, cvar, media, median))
```
